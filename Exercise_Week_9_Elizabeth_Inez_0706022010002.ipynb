**ASSOCIATION RULE SUMMARY**

> 
Data mining menurut Han, adalah proses dasar dimana metode yang cerdas digunakan untuk menghasilkan pola data (Han et al., 2012). Data mining menurut (Sohail, Irshad, Muhammad, Wudil, & Abir, 2019) merupakan tehnik penggalian data untuk menemukan pola yang bisa diklasifikasikan ke dalam pengetahuan yang berguna. Jadi dapat disimpulkan bahwa data mining adalah proses mengekstrak data dengan menggunakan metode dan algoritma tertentu untuk menghasilkan informasi yang lebih berguna sehingga dapat menjadi dasar dalam pengambilan keputusan (Yulianto, 2021). Data mining memiliki beberapa teknik seperti classification, clustering, association rule dan regression.
> 
Analisis asosiasi atau association rule mining adalah teknik data mining untuk menemukan aturan asosiatif antara suatu kombinasi item. Analisis asosiasi dikenal sebagai salah satu teknik data mining yang menjadi dasar dari teknik data mining lainnya (Fauzy, 2015). Singkatnya, association rule mengidentifikasi hubungan dari sekumpulan item yang ada di database. Hubungan ini bukan berdasarkan properti dari data tersebut, tetapi didasarkan dari peristiwa item data. Terdapat beberapa jenis analisis asosiasi, yakni boolean association rules yang digunakan di market basket analysis (MBA), multilevel association rules, qualitative association rules yang diinduksi dari data bisnis, spatial association rules dan lain-lain (Moshkova, Zielosko, & Tetteh, 2022). Metode asosiasi yang umum digunakan adalah FP-Growth, Coefficient of Correlation, Chi Square, A Priori dan lain sebagainya (Muslim, Prasetiyo, & Listiana, 2019).
> 
Aturan asosiasi ini juga disebut market basket analysis. Tujuan dari analisis asosiasi ini yakni untuk menentukan produk yang cenderung dibeli oleh pelanggan (Hadi, 1970). Dengan mengetahui produk yang sering di beli oleh pelanggan, maka pengaturan produk pada supermarket bisa lebih tertata sehingga meningkatkan keuntungan supermarket (Kaur & Kang, 2016). Hal yang sama juga berlaku untuk e-commerce seperti Amazon. Amazon mampu mendeteksi buku mana yang sering dibeli oleh pembeli yang sama. Kemampuan ini membuat Amazon dapat menjual buku dengan lebih efektif menggunakan metode cross-selling (Goedegebuure, 2021). Dalam market basket analysis terdapat set item. Set item adalah kumpulan dari satu atau lebih item. Item ditampilkan dengan kurung kurawal ({}). Contoh kombinasi item adalah sebagai berikut :

Dalam market basket analysis, harus ditentukan nilai support dan confidence seperti di bawah ini: 
>support(A -> B)= P(A U B)


> 


>confidence(A -> B) = P(B | A)
> 
Support merupakan persentase kombinasi item yang dibeli secara bersamaan (A dan B). Sedangkan confidence merupakan persentase kepastian untuk membeli pasangan dari item yang dibeli (jika A maka beli B). Nilai support dan confidence merupakan nilai antara 0-1.

Berikut adalah contoh penerapan association rule pada python :

```python
# Langkah pertama, input data dan temukan set data dalam dataset yang digunakan

data = [
        ['T100',['I1','I2','I5']],
        ['T200',['I2','I4']],
        ['T300',['I2','I3']],
        ['T400',['I1','I2','I4']],
        ['T500',['I1','I3']],
        ['T600',['I2','I3']],
        ['T700',['I1','I3']],
        ['T800',['I1','I2','I3','I5']],
        ['T900',['I1','I2','I3']]
        ]

init = []
for i in data:
    for q in i[1]:
        if(q not in init):
            init.append(q)
init = sorted(init)
print(init)
```

```python
# Langkah kedua, tentukan nilai support. Pada percobaan kali ini kita akan menggunakan support 40%

sp = 0.4
s = int(sp*len(init))
s
```

```python
# Langkah ketiga, kita menerapkan algoritma apriori 

from collections import Counter
c = Counter()
for i in init:
    for d in data:
        if(i in d[1]):
            c[i]+=1
print("C1:")
for i in c:
    print(str([i])+": "+str(c[i]))
print()
l = Counter()
for i in c:
    if(c[i] >= s):
        l[frozenset([i])]+=c[i]
print("L1:")
for i in l:
    print(str(list(i))+": "+str(l[i]))
print()
pl = l
pos = 1
for count in range (2,1000):
    nc = set()
    temp = list(l)
    for i in range(0,len(temp)):
        for j in range(i+1,len(temp)):
            t = temp[i].union(temp[j])
            if(len(t) == count):
                nc.add(temp[i].union(temp[j]))
    nc = list(nc)
    c = Counter()
    for i in nc:
        c[i] = 0
        for q in data:
            temp = set(q[1])
            if(i.issubset(temp)):
                c[i]+=1
    print("C"+str(count)+":")
    for i in c:
        print(str(list(i))+": "+str(c[i]))
    print()
    l = Counter()
    for i in c:
        if(c[i] >= s):
            l[i]+=c[i]
    print("L"+str(count)+":")
    for i in l:
        print(str(list(i))+": "+str(l[i]))
    print()
    if(len(l) == 0):
        break
    pl = l
    pos = count
print("Result: ")
print("L"+str(pos)+":")
for i in pl:
    print(str(list(i))+": "+str(pl[i]))
print()
```

```python
# Langkah keempat, setelah kita mendapatkan final subsets, kita akan mencari asosiasi dari subset-subset yang ada

from itertools import combinations
for l in pl:
    c = [frozenset(q) for q in combinations(l,len(l)-1)]
    mmax = 0
    for a in c:
        b = l-a
        ab = l
        sab = 0
        sa = 0
        sb = 0
        for q in data:
            temp = set(q[1])
            if(a.issubset(temp)):
                sa+=1
            if(b.issubset(temp)):
                sb+=1
            if(ab.issubset(temp)):
                sab+=1
        temp = sab/sa*100
        if(temp > mmax):
            mmax = temp
        temp = sab/sb*100
        if(temp > mmax):
            mmax = temp
        print(str(list(a))+" -> "+str(list(b))+" = "+str(sab/sa*100)+"%")
        print(str(list(b))+" -> "+str(list(a))+" = "+str(sab/sb*100)+"%")
    curr = 1
    print("choosing:", end=' ')
    for a in c:
        b = l-a
        ab = l
        sab = 0
        sa = 0
        sb = 0
        for q in data:
            temp = set(q[1])
            if(a.issubset(temp)):
                sa+=1
            if(b.issubset(temp)):
                sb+=1
            if(ab.issubset(temp)):
                sab+=1
        temp = sab/sa*100
        if(temp == mmax):
            print(curr, end = ' ')
        curr += 1
        temp = sab/sb*100
        if(temp == mmax):
            print(curr, end = ' ')
        curr += 1
    print()
    print()
```
